```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

def f(x):
    if -np.pi <= x<0:
        return x**2 - 1;
    elif 0 <= x <= np.pi:
        return x * np.exp(-x/2)
    
#Plotting the periodic function f(x) over the domain

x = np.linspace(-2*np.pi, 2*np.pi, 400)
y = [f(xi) for xi in x]
plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.title("Periodic Function f(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

def f(x):
    if -np.pi <= x<0:
        return x**2 - 1
    elif 0 <= x <= np.pi:
        return x * np.exp(-x/2)
    else:
        return None  # Handle values of x outside the specified range
    
f_vec = np.vectorize(f)
    
#Plotting the periodic function f(x) over the domain

x = np.linspace(-2*np.pi, 2*np.pi, 1000)

y = f_vec(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y)
plt.title("Periodic Function f(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```

### Question 1b

```python
import numpy as np
from scipy.integrate import quad
import matplotlib.pyplot as plt

# Calculate the coefficients

def fourier_coefficient(n, L):
    a_n = (1 / L) * quad(lambda x: f(x) * np.cos(n * x), -L, L)[0]
    b_n = (1 / L) * quad(lambda x: f(x) * np.sin(n * x), -L, L)[0]
    return a_n, b_n

# Number of harmonics
N = 150

# Period of the function
L = np.pi

# Calculate Fourier coefficients up to N harmonics
a_coeffs = []
b_coeffs = []
for n in range(1, N + 1):
    a_n, b_n = fourier_coefficient(n, L)
    a_coeffs.append(a_n)
    b_coeffs.append(b_n)

# Display the Fourier coefficients (you can print them if needed)
for n in range(1, N + 1):
    print(f'a_{n} = {a_coeffs[n-1]}, b_{n} = {b_coeffs[n-1]}')
# Now, you have the Fourier coefficients for the representation.
```

```python
def a_n(n):
    integral, _ = quad(lambda x: f(x) * np.cos(n * x), -np.pi, np.pi)
    return (1 / np.pi) * integral

def b_n(n):
    integral, _ = quad(lambda x: f(x) * np.sin(n * x), -np.pi, np.pi)
    return (1 / np.pi) * integral

# Define the Fourier series approximation
def fourier_series(x, N):
    result = a_n(0) / 2
    for n in range(1, N + 1):
        result += a_n(n) * np.cos(n * x) + b_n(n) * np.sin(n * x)
    return result
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
```

### Question 1c

```python
def fourier_series(x, N, L, a_coeffs, b_coeffs):
    result = 0.5 * a_coeffs[0]  # Initialize with a0 / 2
    for n in range(1, N + 1):
        result += a_coeffs[n - 1] * np.cos(n * x) + b_coeffs[n - 1] * np.sin(n * x)
    return result

# Plot the Fourier series expansion for various harmonic terms
x_values = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
y_values_1 = [fourier_series(x, 1, L, a_coeffs[:1], b_coeffs[:1]) for x in x_values]
y_values_5 = [fourier_series(x, 5, L, a_coeffs[:5], b_coeffs[:5]) for x in x_values]
y_values_150 = [fourier_series(x, 150, L, a_coeffs, b_coeffs) for x in x_values]

plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values_1, label='1st Harmonic')
plt.plot(x_values, y_values_5, label='5th Harmonic')
plt.plot(x_values, y_values_150, label='150th Harmonic')
plt.title("Fourier Series Expansion")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
